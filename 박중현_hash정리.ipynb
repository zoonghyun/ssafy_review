```
#include <unordered_map>
unordered_map<int, Node> um;
```


데이터 추가, 수정, 삭제, 검색

```
    // data 추가
    um.insert({ 1, {5, 3} });
    um.insert({ 5678, {12, 9} });
    um.insert({ 9, {21, 4} });
    um.insert({ 7, {8, 9753} });
    um.insert({ 0, {1910, 32} });
    um.insert({ -2134567890, {2134567890, 1234567898} });

    // data 수정 or 추가
    um[1] = { 951, 7753 }; // 1번 key의 value를 수정
    um[123] = { 987, 4 }; // 없는 key의 경우 추가됨

    // data 삭제
    um.erase(1); // 해당 key와 value의 데이터 삭제

    // data 찾기
    // -특정 key의 iterator 가져오기
    auto it_minus = um.find(-2134567890);
    // 응용 : 특정 key가 존재?
    if (um.find(1) == um.end()) // 찾지 못하면 end가 반환됨.
        cout << "찾지 못함" << endl;
    else
        cout << "찾음" << endl;
```



해시 함수


```
struct HashFunction_return1 {
  size_t operator()(int key) const {
    return 1; // 해시 충돌 발생. O(N)
  }
};

struct HashFuncNode {
  size_t operator()(Node key) const {
    return key.row * 10000 + key.col; // 충돌 최대한 방지.
  }
};

struct HashFuncVector {
  size_t operator()(vector<int>key) const {
    size_t ret = 0;
    for (auto el : key )
      ret += el;
    return ret;
  }
};
```
