- Vue.js는 개발자가 웹 애플리케이션을 개발하고 유지 관리하는 데 도움을 주는 JavaScript 프레임워크
    
    → 웹에 배포할 때는 HTML, CSS 및 JavaScript 형식으로 변환된 후 사용자에게 제공됨
- Vue.js 프로젝트는 Node.js의 npm을 통해 라이브러리 및 의존성 관리 (package.json 생성)
    
    → Node.js 설치 필요

- 컴포넌트 단위로 나누어 개발

# **프로젝트 생성**

===================================================
Vue.js devtools 설치 (크롬 확장 프로그램)
  
  → Vue.js로 개발한 웹 애플리케이션의 디버깅 및 개발 지원

  기능 :

  1. **컴포넌트 트리 검사**: Vue.js 애플리케이션의 컴포넌트 트리를 시각적으로 확인
  2. **실시간 데이터 변경 모니터링**: 애플리케이션의 상태 데이터를 모니터링 및 실시간으로 변경 사항을 확인
  3. **컴포넌트 상태 검사**: 각 컴포넌트의 상태 및 속성을 검사 및 변경 가능
  4. **이벤트 트래킹**: 이벤트 핸들러 추적을 통한 이벤트 발생 확인

  =====================================================

  - 프로젝트 폴더에서 터미널 생성 후 `**npm init vue@3**` 명령어 입력
    - npm을 사용하여 Vue.js 3 프로젝트 초기화
    - package.json 파일을 만들기 위한 질문에 응답 필요
    
      ⇒ 프로젝트 초기화 및 기본 설정 생성을 자동화

      1. Vue Router : Vue.js 애플리케이션에서 네비게이션을 처리하기 위해 필요

      - 네비게이션이란?
        - **라우팅**: 사용자가 웹 애플리케이션에서 특정 URL을 입력 또는 링크 클릭 시 해당 페이지로 이동할 수 있도록 하는 기능
        - **URL 관리**: 사용자가 특정 URL로 이동 시 애플리케이션은 해당 URL에 대한 적절한 내용을 표시. 뒤로 가기 또는 앞으로 가기 버튼 클릭 시에도 URL을 기반으로 페이지 이동 및 상태 관리
        - **상태 관리**: 이전 페이지의 정보를 다음 페이지로 전달하거나 페이지 간에 데이터를 공유하는 등의 작업을 수행

      2. Pinia : 상태 관리와 관련된 작업을 단순화하고 효율적으로 처리할 수 있도록 도와주는 라이브러리

- `**npm install**` 명령어 입력
    - `**npm init vue@3**` 명령어는 프로젝트의 종속성 패키지 (dependencies)를 설치하지는 않음
    - dependencies 설치
      - dependency란?
        패키지를 실행시키기 위해 필요한 외부 라이브러리, 모듈, 도구 등의 "다른 패키지" 목록으로 package.json에서 명시

      - 다음 기능 등을 제공
      1. **라이브러리 및 모듈 사용**: 다른 개발자들이 개발한 라이브러리 및 모듈을 프로젝트에서 활용할 수 있도록 함
      2. **플러그인 및 확장 기능**: 특정 기능을 추가하거나 확장하기 위해 사용.
        *ex) Vue Router 라이브러리 추가를 통한 라우팅 기능 확장*
      3. **개발 도구 및 빌드 도구**: 프로젝트의 개발 및 빌드 프로세스를 지원하는 도구와 라이브러리 포함
      4. **스타일 및 리소스 관리**: CSS 프레임워크, 이미지 처리 도구, 데이터베이스 라이브러리 등과 같은 리소스와 스타일 관리에 사용

- **`npm run dev`** 명령어 입력
    - 개발 서버 실행
        - 웹 애플리케이션을 로컬 환경에서 실행하고 브라우저에서 테스트할 수 있는 환경을 제공
        - 코드 수정 및 변경 사항 실시간 감지 및 즉시 다시 빌드  & 브라우저 자동 새로 고침. 빠른 개발 및 디버깅 가능
